Api_key (período de 01-01-2023 à 01-01-2024 -> 1 ano)

```python
url = "https://olinda.bcb.gov.br/olinda/servico/PTAX/versao/v1/odata/CotacaoDolarPeriodo(dataInicial=@dataInicial,dataFinalCotacao=@dataFinalCotacao)?@dataInicial='01-01-2023'&@dataFinalCotacao='01-01-2024'&$top=1000&$skip=0&$format=json&$select=cotacaoCompra,cotacaoVenda,dataHoraCotacao"
```

Importação das Bibliotecas

```python
import pandas as pd  #possibilita manipulação e análise de dados
import requests  #possibilita integração com serviços web, solicitações HTTP consumir dados de APIs
import json  #manipular arquivo json
```

Função para extrair os dados da API

```python
def requisicao_api(link):
  resposta = requests.get(link)

  if resposta.status_code == 200:      #sucesso na requisição
    dados = resposta.json()
    print('Status Code:', resposta.status_code)
    with open('resultado_Periodo.json', 'w', encoding='utf-8') as arquivo:   #salvar arquivo json
      json.dump(dados, arquivo, ensure_ascii=False, indent=4)

  else:
    print('Status Code:', resposta.status_code)
```

```python
requisicao_api(url)
```

Conversão em dataframe (para Json)

```python
jsondata = pd.read_json('resultado_Periodo.json')
```

```python
data = jsondata['value']
dados_df = pd.json_normalize(data)
```

Exportando para CSV

```python
df = pd.json_normalize(jsondata['value']) ; 
df.to_csv("resultado_Periodo.csv", index=False)
```
